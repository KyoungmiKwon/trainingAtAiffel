```python
# 조회
## attribute
## 타입변환 (astype)

# 정렬(sort)
df.sort_index().head() # 인덱스 정렬
df.sort_index(ascending=False).head() # 인덱스 내림차순 정렬

df.sort_values(by='age').head() # 값 정렬
df.sort_values(by='age', ascending=False).head() # 값 내림차순 정렬
df.sort_values(by=['age', 'fare']).head() # 다중기준 정렬
df.sort_values(by=['age', 'fare'], ascending=[True, False]).head() # 다중기준 정렬, with 정렬방식 지정

# 조건필터, indexing, slicing
```

```python
# 조회
## attribute
## 타입변환 (astype)

# 정렬(sort)

# indexing, slicing, 조건필터

## loc
df.loc[5,'class'] # indexing
df.loc[2:5, ['age', 'fare','who']] # fancy indexing [시작(포함):끝(포함)]

df.loc[2:5, 'class':'deck'] # slicing [시작(포함):끝(포함)]
df.loc[:6, 'class':'deck'] # slicing [:끝(포함)]

cond = (df['age'] >= 70) # boolean indexing
cond
df.loc[cond].head()

df.loc[(df['age'] >=70),'class':'deck'] # boolean slicing

### 다중조건
cond1 = (df['fare'] <= 30)
cond2 = (df['who'] == 'woman')

df.loc[cond1 & cond2] # and
df.loc[(df['fare'] <= 30) & (df['who'] == 'woman')]

df.loc[cond1 | cond2] # or
df.loc[(df['fare'] <= 30) | (df['who'] == 'woman')]

df.loc[(df['age'] >= 70),'age'] # boolean slicing
df.loc[(df['age'] >= 70),'age'] = -1 # 조건 필터 후 원하는 값 대입 가능 : 단일 컬럼 선택에 유의
df.loc[(df['age'] == -1)] # 확인


## iloc ; index 기반, indexing, slicing 모두 가능
df.iloc[1,3] # indexing
df.iloc[[1,3,5], [2,4,5,6,9]] # fancy indexing
df.iloc[:3, :5] # slicing
df.iloc[1:3, 2:5] # slicing

## 조건필터
### isin : 특정 값의 포함 여부 비교
df['who'].isin(['child'])
df.isin(['child'])

condi = df['who'].isin(['child']) #조건 필터링
df.loc[condi]
```
