```python
'''groupby() / pivot_table() / concat() / merge()

groupby() : 데이터를 특정 조건에 따라 피봇팅 하여 통계량 확인
pivot_table() : 데이터를 특정 조건에 따라 행과 열을 기준으로 펼쳐 통계량 확인
concat() : 2개 이상의 데이터 프레임을 행 혹은 열 방향으로 연결
merge() : 2개의 데이터 프레임을 특정 Key를 기준으로 병합
'''
# apply(function 함수 / lambda 함수) ; 복잡한 논리연산을 컬럼 또는 데이터 프레임에 적용하고자 할 때 사용
## function 함수
## lambda 함수

# groupby()
## reset_index() : 그룹핑 된 데이터 프레임의 index를 초기화 하여 새로운 데이터 프레임을 생성

# concat([데이터프레임1, 데이터프레임2], axis = 0/1)
## axis = 0 : 행 방향으로 연결
### 같은 column을 알아서 찾아서 데이터를 연결
### 연결시 index가 초기화 되지 않아, 데이터프레임의 갯수와 index가 맞지 않음 => ignore_index = True
### 합치고자 하는 데이터 프레임으 일부 컬럼이 누락되거나 순서가 바뀌어도 알아서 같은 칼럼끼리 병합

## axis = 1 : 열 방향으로 연결
### 같은 index 행끼리 연결됨

# merge()
## 다른 구성의 데이터프레임이지만, 공통된 컬럼값을 가지고 있다면 병합할 수 있음
## 병합하는 방법 4가지 - how : {left, right, outer, inner} , inner가 기본값
## 병합하려는 칼럼의 이름이 다른 경우
### pd.merge(데이터프레임1,데이터프레임2, left_on = 데이터프레임1의 컬럼이름 , right_on = 데이터프레임2의 컬럼이름)

# 데이터 셋
df = sns.load_dataset('titanic')
df.head() 

gas1 = pd.read_csv('/Users/kwonkyoungmi/workspace_Python/data/유가정보/gas_first_2019.csv', encoding = 'euc-kr')
gas1.head()
```

```python
# who 컬럼에 대하여 man은 남자, woman은 여자, child는 아이로 변경 하기

# 1. category면 (로 변경하여)rename

df.who_type.info()

df['who_type'] = df['who'].astype('category') #dtype 을 catetory로 변경
df.who_type.cat.categories # category array 순서 확인
df['who_type'] = df['who_type'].cat.rename_categories(['아이', '남자', '여자']) # category 이름 변경

df[['who','who_type']]
print(df.who.value_counts(), df.who_type.value_counts())

# 2. apply() 활용
## function 정의 하여 변환 하기
def transform_who(x):
    if x == 'man':
        return '남자'
    elif x == 'woman':
        return '여자'
    else :
        return '아이'
df.who.apply(transform_who).value_counts()    

## lambda 함수 정의 하여 변환하기 : 간단한 논리연산의 경우 
## -- 0 : 사망, 1 : 생존 으로 변경
print(df.survived.value_counts())
df.survived.apply(lambda x : '생존' if x == 1 else '사망').value_counts()
```

```python
'''groupby()'''
# 생존자와 사망자를 [성별을 기준]으로 그룹핑 하여 평균 살펴보기
# 통계함수와 일반적으로 같이 적용

df.groupby('sex').mean(numeric_only=True)
```

```python
df.groupby(['sex','pclass']).mean(numeric_only=True) # 2개의 이상의 컬럼으로 그룹핑
```

```python
df.groupby(['sex','pclass']).survived.mean() # survived 컬럼에 대한 결과만 도출
```

```python
pd.DataFrame(df.groupby(['sex','pclass']).survived.mean()) # 데이터 프레임으로 출력
```

```python
pd.DataFrame(df.groupby(['sex','pclass']).survived.mean()).reset_index() # 인덱스 초기화
```

```python
df.groupby(['sex','pclass'])[['survived','age']].mean() # 다중 컬럼 지정
```

```python
df.groupby(['sex','pclass'])[['survived','age']].agg(['mean','sum']) # 다중 통계 함수 적용
```

```python
'''pivot()'''
# index, columns, values 지정하여 피벗

df.pivot_table(index = 'who', values = 'survived')  # 인덱스에 그룹을 표기
```

```python
df.pivot_table(columns = 'who', values = 'survived') # 컬럼에 그룹을 표기
```

```python
df.pivot_table(index = ['who','pclass'], values = 'survived') # 다중 그룹에 대한 단일 컬럼 결과
```

```python
df.pivot_table(index = 'who', columns = 'pclass', values = 'survived') # 인덱스에 중첩하지 않고 행과 열
```

```python
df.pivot_table(index = 'who', columns = 'pclass', values = 'survived', aggfunc = ['sum','mean']) # 다중 통계함수 적용
```

```python
'''merge()'''
# 다른 구성의 데이터프레임이지만, 공통된 컬럼값을 가지고 있다면 병합할 수 있음


df1 = pd.DataFrame({
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']})
print(df1)

df2 = pd.DataFrame({
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']})
print("");print(df2)

pd.merge(df1,df2)
```

```python
# 병합하는 방법 4가지 - how : {left, right, outer, inner} , inner가 기본값

print(pd.merge(df1,df2))
print('');print(pd.merge(df1,df2, how='left'))
print('');print(pd.merge(df1,df2, how = 'right'))
print('');print(pd.merge(df1,df2, how = 'outer'))
```

```python
# 병합하려는 칼럼의 이름이 다른 경우
df3 = pd.DataFrame({
    '이름': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']})
print(df3)

df4 = pd.DataFrame({
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']})
print(df4)
```

```python
pd.merge(df3,df4, left_on = '이름', right_on = '고객명') # 고객명이 그대로 있음
```
