### 1. Scartter Plot

```python
# Scartter Plot

x = np.random.rand(50)
y = np.random.rand(50)

colors = np.arange(50)  # 점의 색 :  랜덤
area = x * y * 250      # 점의 크기 : x,y 값이 커지면 넓이가 커지도록 설정
plt.scatter(x, y, s=area, c=colors, alpha=0.5) # alpha : 투명도

plt.show()
```

### 2. Barplot, Barhplot

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
# Barplot
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]
plt.bar(x,y, align='center', alpha = 0.5, color='blue', width= .5) # width = 막대의 굵기 조정
plt.xticks(x, fontsize = 8)
plt.ylabel('number of student')
plt.title('Subjects', color = 'blue')

plt.subplot(1,2,2)
# Bar'h'plot
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]
plt.barh(x, y, align='center', alpha=0.3, color='green') # plt.barh
plt.yticks(x, fontsize = 8, color ='red', alpha = .5)    # xticks로 설절했던 부분을 yticks로 변경
plt.xlabel('Number of Students', color = 'green')
plt.title('축 변환한 자료', color = 'green')

plt.show()

```

```python
# Barplo에서 비교 그래프 그리기

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
# single Barplot
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]

plt.bar(x,y, align='center', alpha = 0.5, color='blue', width= .5) # width = 막대의 굵기 조정
plt.xticks(x, fontsize = 8, rotation=45)  # Added rotation for readability
plt.ylabel('number of student')
plt.title('Single Barplot', color = 'blue')



plt.subplot(1,2,2) # 1 row, 2 columns, second subplot
# Comparative Bar plot
x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y_1 = [66, 80, 60, 50, 80, 10]
y_2 = [55, 90, 40, 60, 70, 20]

width = 0.35 ; x = np.arange(len(x_label)) # Bar 넓이 설정 위해서

plt.bar(x - width/2, y_1, width, align='center', alpha=0.5, label= 'John')
plt.bar(x + width/2, y_2, width, align='center', alpha=0.8, label= 'Peter')

plt.xticks(x, x_label, rotation = 45) # set xtick & label
plt.ylabel('Number of Students')
plt.title('Comparative Bar')
plt.legend(['john', 'peter'])

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
# 축 변환
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
# Comparative Bar plot
x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y_1 = [66, 80, 60, 50, 80, 10]
y_2 = [55, 90, 40, 60, 70, 20]

width = 0.35; x = np.arange(len(x_label))
plt.bar(x - width/2, y_1, width, align='center', alpha=0.5, label='John') #x - width/2: 막대를 왼쪽으로 약간 이동
plt.bar(x + width/2, y_2, width, align='center', alpha=0.8, label='Peter')

plt.xticks(x, x_label, rotation=45)  # Set x-ticks and labels
plt.ylabel('Number of Students')
plt.title('General')
plt.legend()  # Show legend



plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
# Axis translation
x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y_1 = [66, 80, 60, 50, 80, 10]
y_2 = [55, 90, 40, 60, 70, 20]

width = 0.35; x = np.arange(len(x_label))
plt.barh(x - width/2, y_1, width, align='center', alpha=0.5, label='John') #x - width/2: 막대를 왼쪽으로 약간 이동
plt.barh(x + width/2, y_2, width, align='center', alpha=0.8, label='Peter')

plt.yticks(x, x_label, rotation=45)  # Set x-ticks and labels
plt.xlabel('Number of Students')
plt.title('Axis translation')
plt.legend()  # Show legend

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

### 3. Line Plot

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
# Single line plot
x = np.arange(0, 10, 0.1)
y = 1 + np.sin(x)

plt.plot(x, y)

plt.xlabel('x value', fontsize=10)
plt.ylabel('y value', fontsize=10)
plt.title('sin graph', fontsize=9)

plt.grid(alpha = .3)


plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
# multi line plot
x = np.arange(0, 10, 0.1)
y_1 = np.sin(x)
y_2 = np.cos(x)

plt.plot(x, y_1, label='sin', color='c', alpha=0.5, marker = '*', linewidth = 10, linestyle = ':')
plt.plot(x, y_2, label='cos', color='red', alpha=0.7, marker = "+")

plt.xlabel('x value', fontsize=15)
plt.ylabel('y value', fontsize=15)
plt.title('sin and cos graph', fontsize=9)

plt.grid(alpha = 0.1, color = 'green')
plt.legend()

plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()


```

### 4. Areaplot (Filled Area)

```python
plt.figure(figsize=(18,5))

plt.subplot(1, 3, 1)
# 기본 area plot
x = np.arange(1,21)
y =  np.random.randint(low=5, high=10, size=20)

## fill_between으로 색칠하기
plt.fill_between(x, y, color="green", alpha=0.5)
plt.title('basic area plot')

plt.subplot(1, 3, 2)
# 경계선 굵게 그리고 area는 옅게 그리는 효과 적용
x = np.arange(1,21)
y =  np.random.randint(low=5, high=10, size=20)

plt.fill_between(x, y, color="green", alpha=0.3)
plt.plot(x, y, color="green", alpha=0.8, linewidth = 3) # 경계선 굵게 그리기
plt.title('Thick Line')


plt.subplot(1, 3, 3)
# 여러 그래프 겹쳐 그리기
x = np.arange(1, 10, 0.05)
y_1 =  np.cos(x)
y_2 =  np.sin(x)
y_3 = (y_1 + 3/4) * y_2 / np.pi

plt.fill_between(x, y_1, color="green", alpha=0.1)
plt.fill_between(x, y_2, color="blue", alpha=0.2)
plt.fill_between(x, y_3, color="red", alpha=0.3)
plt.title('over lap')



plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

### 5. Histogram

```python
# Basic histogram

N = 100000
x = np.random.randn(N)
bins = 30

# bin의 크기 변화
fig, axs = plt.subplots(1,3,
                        sharey = True,      # y축을 다중 그래프가 share
                        tight_layout = True
                        )

fig.set_size_inches(12, 3)

axs[0].hist(x, bins = bins)
axs[1].hist(x, bins = bins*2)
axs[2].hist(x, bins = bins*4)
plt.show()


# Y축에 Density 표시 하기
fig, axs = plt.subplots(1,2,
                        tight_layout = True
                        )

fig.set_size_inches(12, 3)

axs[0].hist(x, bins = bins, density = True, cumulative = True)
axs[1].hist(x, bins = bins, density = True)
# axs[1,2].hist(x, bins = bins, cumulative = True)
plt.show()
```

### 6. Pie Chart

```python
''' pie chart 옵션
explode: 파이에서 툭 튀어져 나온 비율
autopct: 퍼센트 자동으로 표기
shadow: 그림자 표시
startangle: 파이를 그리기 시작할 각도
texts: label에 대한 텍스트 효과
autotexts: 파이 위에 그려지는 텍스트 효과를 다룰 때 활용
'''

plt.figure(figsize=(13,8))

labels = ['Samsung', 'Huawei', 'Apple', 'Xiaomi', 'Oppo', 'Etc']
sizes = [20.4, 15.8, 10.5, 9, 7.6, 36.7]
explode = (0.3, 0, 0, 0, 0, 0)

# texts, autotexts 인자를 활용하여 텍스트 스타일링을 적용
patches, texts, autotexts = plt.pie(sizes, 
                                    explode=explode, 
                                    labels=labels,  
                                    autopct='%1.1f%%',
                                    shadow=True, 
                                    startangle=90,
                                    #hatch = ['**0','oO','O.O','.||.']  # 패턴
                                    #pctdistance=1.25,
                                    #labeldistance= .6
                                    )


plt.title('Smartphone pie', fontsize=15)

# label 텍스트에 대한 스타일 적용
for t in texts:
    t.set_fontsize(12)
    t.set_color('gray')

# pie 위의 텍스트에 대한 스타일 적용
for t in autotexts:
    t.set_color("white")
    t.set_fontsize(10)

plt.legend(title = 'companies ;')
plt.show()
```

### 7. Box Plot

```python
# 샘플 데이터 생성
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))
```

```python
fig, axs = plt.subplots(1,3,
                        tight_layout = True
                        )

fig.set_size_inches(12, 6)

axs[0].boxplot(data)
axs[0].set_title('Basic Box Plot')
axs[1].boxplot(data, vert = False)
axs[1].set_title('Horizontal Box Plot', fontsize = 15, color = 'm')
axs[2].boxplot(data, flierprops = dict(markerfacecolor='c', marker='D'))
axs[2].set_title('Changed Outlier Symbols', fontsize = 15, color = 'c')
plt.show()
```

### 8. 3D 그래프

```python
fig = plt.figure()
ax = plt.axes(projection='3d') # project=3d로 설정합니다

# x, y, z 데이터를 생성합니다
z = np.linspace(0, 15, 1000)
x = np.sin(z)
y = np.cos(z)

ax.plot3D(x, y, z, 'gray', alpha=0.03, marker='o')
plt.show()
```
