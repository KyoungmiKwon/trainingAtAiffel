### 연습문제 2.1.1 ; 행렬

```python
'''MNIST 숫자이미지 불러오기
이 데이터는 0~9까지의 숫자 이미지를 8x8 픽셀 해상도의 흑백 이미지로 구성된 것
'''

from sklearn.datasets import load_digits # MNIST 데이터셋을 사용하기 위해 불러옴

digits = load_digits() # MNIST 데이터셋을 불러옴
samples = [0, 10, 20, 30, 1, 11, 21, 31] # 선택된 이미지 번호
```

```python
# 이미지 데이터 셋에서 뽑아오기
d = []

for i in range(8):
    d.append(digits.images[samples[i]]) # 이미지를 d에 저장

# 이미지 출력
plt.figure(figsize=(8, 2)) # 이미지의 크기를 설정
for i in range(8):
    plt.subplot(1,8,i+1) # 1행 8열로 이미지를 출력
    plt.imshow(d[i], cmap=plt.cm.bone_r, interpolation='nearest') # 이미지를 출력
    plt.grid(False) # 이미지의 격자를 출력하지 않음
    plt.xticks([]) # x축의 눈금을 출력하지 않음
    plt.yticks([]) # y축의 눈금을 출력하지 않음
    plt.title("image {}".format(i+1)) # 이미지의 제목을 출력
plt.suptitle("숫자 이미지 8개", fontname = 'NanumGothic') # 이미지의 전체 제목을 출력
plt.tight_layout() # 이미지의 간격을 설정
plt.show()
```

```python
# 2차원 이미지 데이터를 1차원 벡터로 변환
v = []

for i in range(8):
    v.append(d[i].reshape(64,1)) # 이미지를 벡터로 변환하여 v에 저장 (8x8 해상도 이미지 스캔 =64)

# 벡터 출력
plt.figure(figsize=(12, 6)) # 이미지의 크기를 설정
for i in range(8):
    plt.subplot(1,8,i+1) # 1행 8열로 이미지를 출력
    plt.imshow(v[i], aspect=0.4, cmap=plt.cm.bone_r, interpolation='nearest') # 이미지를 출력
    plt.grid(False) # 이미지의 격자를 출력하지 않음
    plt.xticks([]) # x축의 눈금을 출력하지 않음
    plt.yticks([]) # y축의 눈금을 출력하지 않음
    plt.title("벡터 {}".format(i+1), fontname = 'NanumGothic') # 이미지의 제목을 출력
plt.suptitle("벡터화된 이미지 8개", fontname = 'NanumGothic') # 이미지의 전체 제목을 출력
plt.tight_layout() # 이미지의 간격을 설정
plt.show()

# 이미지가 닮았나? 같은 숫자에 대한 벡터가 닮았음에 주의하라는데.....
```

```python
# 유사도
''' 
벡터의 내적은 두 벡터간의 유사도를 계산하는데 사용 될 수 있음
두 벡터가 비슷할수록 내적값이 커짐
cosine similarity : 두 벡터간의 코사인 각도를 이용하여 유사도를 계산
'''
import matplotlib.gridspec as gridspec

df = digits.copy()

d1 = df['images'][0]
d2 = df['images'][10]
d3 = df['images'][1]
d4 = df['images'][11]  # 이미지 4개 선택
d4 = df['images'][11]  # 이미지 4개 선택
v1 = d1.reshape(64,1)
v2 = d2.reshape(64,1)
v3 = d3.reshape(64,1)
v4 = d4.reshape(64,1)  # 이미지를 벡터로 변환

plt.figure(figsize=(9,9))
gs = gridspec.GridSpec(1,8, height_ratios=[1], width_ratios=[9,1,9,1,9,1,9,1])  # 그리드 생성

for i in range(4):
    plt.subplot(gs[2*i])  # 이미지 출력
    plt.imshow(eval("d"+str(i+1)), aspect=1, cmap=plt.cm.bone_r, interpolation='nearest') # eval() 함수는 문자열로 작성된 코드를 실행하여 그 결과를 반환
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.title("image {}".format(i+1))

    plt.subplot(gs[2*i+1])  # 벡터 출력
    plt.imshow(eval("v"+str(i+1)), aspect=0.25, cmap=plt.cm.bone_r, interpolation='nearest')
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])
    plt.title("vector {}".format(i+1))

plt.tight_layout()
plt.show()
```

### 연습문제 2.2.2 : 가중평균

벡터 $x$의 평균 제거 벡터는 다음과 같이 계산함을 증명하라.

$$ 
\begin{align}
x - \dfrac{1}{N}\mathbf{1}_N^Tx \mathbf{1}^{}_N
\tag{2.2.32}
\end{align}
$$

x = vector  
N은 원소 갯수

```python
'''
x - 1/N  * 1_N^T *  x  * 1_N
== x - mean(x) * 1_N
== x - mean(x)
'''

# 가중평균
x = np.arange(10)
N = len(x)

N_1 = np.ones(N)  # 1로 이루어진 벡터 ; 1_N
N_1_T = N_1.T     # 1_N^T

print(x - np.dot(np.dot( N_1_T, x), N_1) / N)
print(np.mean(x))
print(x - np.mean(x))
```

### 2.2.12 : 이차형식

다음 식이 성립함을 증명하라.

$$ 
x^TAx = \dfrac{1}{2}x^T(A + A^T)x 
$$

주어진 식은 행렬 $A$가 대칭 행렬이 아니더라도 성립.     
왜냐하면 $x^TAx$와 $x^TA^Tx$의 합은 대칭적인 형태를 가지기 때문     
$a_{ij}x_ix_j$와 $a_{ji}x_jx_i$는 서로 같은 항이므로, 이들의 합은 항상 대칭적인 형태        
>>    $\dfrac{1}{2}x^T(A + A^T)x$는   
>>    항상 $x^TAx$와 같음       

## 4. 행렬의 성질
1. 정부호와 준정부호
     - 양의 정부호(positive definite) : $$x^T A x > 0$$
     - 양의 준정부호(positive semi-definite) : $$ x^T A x \geq 0 $$

2. 행렬 놈
     - 놈의 정의 : 다음 4가지 성질이 성립하는 행렬 연산


          - 놈의 값은 0이상이다. 영행렬일 때만 놈의 값이 0이 된다.  $$\|A\|\ge 0$$
          - 행렬에 스칼라를 곱하면 놈의 값도 그 스칼라의 절대값을 곱한 것과 같다.   $$\|\alpha A\|=|\alpha| \|A\|$$
          - 행렬의 합의 놈은 각 행렬의 놈의 합보다 작거나 같다.   $$\|A+B\| \le \|A\|+\|B\|$$
          - 정방행렬의 곱의 놈은 각 정방행렬의 놈의 곱보다 작거나 같다.  $$\|AB\| \le \|A\|\|B\|$$

3. 대각합 : 정방행렬에서만 정의 되며 대각원소의 합으로 계산

     - 대각합의 성질 ($c$는 스칼라, $A, B, C$는 행렬)

          - 스칼라를 곱하면 대각합은 스칼라와 원래의 대각합의 곱이다.   $$ \text{tr} (cA) = c\;\text{tr} (A) $$

          - 전치연산을 해도 대각합이 달라지지 않는다.       $$ \text{tr} (A^T) = \text{tr} (A)  $$

          - 두 행렬의 합의 대각합은 두 행렬의 대각합의 합이다.   $$ \text{tr} (A + B) = \text{tr} (A) + \text{tr} (B)  $$

          - 두 행렬의 곱의 대각합은 행렬의 순서를 바꾸어도 달라지지 않는다.     $$\text{tr} (AB) = \text{tr} (BA)  $$

          - 세 행렬의 곱의 대각합은 다음과 같이 순서를 순환시켜도 달라지지 않는다.        $$ \text{tr} (ABC) = \text{tr} (BCA) = \text{tr} (CAB)  $$

     - 곱셈에서 순서가 바뀌어도 대각합이 같다는 것을 이용하여 아래의 식 도출    $$\text{tr} ((AB)C) = \text{tr} (C(AB)) = \text{tr} ((CA)B) = \text{tr} (B(CA)) $$

          - 트레이스 트릭(trace trick) 
               >> 이차 형식(quadratic form)의 미분을 구하는 데 유용하게 사용    

               >> 이 두 식에서는 A, B, C가 각각 정방행렬일 필요는 없고, 최종적으로 대각합을 구하는 행렬만 정방행렬이기만 하면 됨
          - 이차형식의 트레이스 트릭 
          $$x^TAx = \text{tr}(x^TAx) = \text{tr}(Axx^T)  = \text{tr}(xx^TA)  $$

4. 행렬식

     - 행렬식의 성질

          - 전치 행렬의 행렬식은 원래의 행렬의 행렬식과 같다.    $$ \det(A^{T}) = \det(A)  $$
          - 항등 행렬의 행렬식은 1이다. $$ \det(I) = 1  $$
          - 두 행렬의 곱의 행렬식은 각 행렬의 행렬식의 곱과 같다.     $$ \det(AB) = \det(A)\det(B)  $$


     -  행렬식의 계산
          -  2×2 행렬의 행렬식          $$ \det \left( \begin{bmatrix}a&b\\c&d\end{bmatrix} \right) = ad-bc $$

          -  3×3 행렬의 행렬식          $$ \det \left( \begin{bmatrix}a&b&c\\d&e&f\\g&h&i\end{bmatrix} \right) =aei+bfg+cdh-ceg-bdi-afh$$



     - **역행렬**        
          >> $A^{-1}$은 원래의 행렬 $A$와 아래의 관계를 만족하는 정방행렬  
          >> $I$는 항등 행렬(identity matrix)이다.      $$ A^{-1} A = A A^{-1} = I  $$

     - 역행렬의 행렬식은 원래의 행렬의  행렬식의 역수와 같음     $$ \det(A^{-1}) = \dfrac{1}{\det(A)}  $$

### 연습문제 2.3.7 : 행렬식

다음 행렬이 양의 정부호인지, 양의 준정부호인지 혹은 두가지 중 어느것도 아닌지 판단하라. 그리고 행렬의 대각합과 행렬식을 구하라.

(1)

$$
\begin{bmatrix} 
2 & -1 & 0 \\ -1 & 2 & -1 \\ 0 & -1 & 2 
\end{bmatrix} $$

(2)

$$
\begin{bmatrix} 
1 & 2 \\ 3 & 4
\end{bmatrix} 
$$


```python
A = np.array([[2, -1, 0], 
              [-1, 2, -1], 
              [0, -1, 2]])

x = np.zeros((3,1))

x.T @ A @ x
```

```python
A = np.array([[2, -1, 0], 
              [-1, 2, -1], 
              [0, -1, 2]])
print(A)
print("대각합 : ",np.trace(A))
print("행렬식 : ", np.linalg.det(A))


B = np.array([[1, 2],
              [3, 4]])

print(B)
print("대각합 : ",np.trace(B))
print("행렬식 : ", np.linalg.det(B))

```
