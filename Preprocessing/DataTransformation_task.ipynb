```python
# Q. car_df와 brand_df를 합치기 위해, car_df에 brand 컬럼을 추가합니다. 
# title 값의 첫 번째 단어만 추출하여 brand 컬럼에 저장해주세요.

#print(np.sort(car['title'].unique()))
#print(np.sort(brand['title'].unique()))

print(brand['title'].nunique())
```

```python
# Q. car_df의 brand 컬럼과 brand_df의 title 컬럼을 기준으로 데이터를 합칩니다.

m_df = c_df.merge(brand, how='left', left_on='brand', right_on='title')
m_df.head(3)
```

```python
# Q. car_df에서 중복되는 컬럼 title_y를 제거합니다.
m_df.drop(columns='title_y', inplace=True)

# Q. car_df에서 title_x 컬럼의 이름을 title로 변경합니다.
m_df.rename(columns={'title_x': 'title'}, inplace=True)

# Q. car_df에서 brand 컬럼을 다시 대문자로 변경 ㅋㅋ
m_df['brand'] = m_df['brand'].str.upper()

m_df.head(3)
```

```python
# 범주형 컬럼들에 원-핫 인코딩을 적용합니다.

s_df = r_df.copy()
s_df = pd.get_dummies(s_df, drop_first = True)
s_df.head(3)
```

```python
# Q. Robust Scaler를 rs 변수로 정의해줍니다.
rs = RobustScaler()
```

```python
# Q. robust scaling을 적용한 데이터프레임을 car_df 변수에 다시 저장합니다.
r_df = s_df.copy()
r_df = rs.fit_transform(s_df)
```

```python
# Q. 주성분이 5개인 PCA 모델을 정의합니다.

pca = PCA(n_components=5)
```

```python
# Q. car_df 데이터에 PCA를 적용합니다.

pca.fit(r_df)
```

```python
# Q. 원본 데이터 대비 PCA 데이터의 분산(explained variance) 비율을 계산합니다.

pca.explained_variance_ratio_
```

```python
# 주성분의 개수가 2~10개일 때 explained variance ratio를 각각 계산합니다.

for i in range(2, 11):
    pca = PCA(i)
    pca.fit(r_df)
    print(i, round(pca.explained_variance_ratio_.sum(), 2))
```
